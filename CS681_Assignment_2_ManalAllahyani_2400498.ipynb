```python
# diabetes_medlineplus_scraper.py
import requests
from bs4 import BeautifulSoup
import json

url = "https://medlineplus.gov/diabetestype2.html"

headers = {"User-Agent": "Mozilla/5.0"}
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, "lxml")

# Grab all visible paragraphs and list items
paragraphs = [p.get_text(" ", strip=True) for p in soup.find_all(["p", "li"])]

# Separate content using keyword clues
symptoms = [t for t in paragraphs if any(k in t.lower() for k in ["symptom", "sign", "thirst", "urination", "fatigue", "blurred"])]
risk_factors = [t for t in paragraphs if any(k in t.lower() for k in ["risk", "overweight", "family history", "obese", "inactive"])]
treatment_options = [t for t in paragraphs if any(k in t.lower() for k in ["treat", "manage", "medicine", "diet", "exercise", "insulin"])]

# Build the knowledge base
knowledge_base = {
    "disease_name": "Type 2 Diabetes Mellitus",
    "symptoms": symptoms[:10],
    "risk_factors": risk_factors[:10],
    "treatment_options": treatment_options[:10],
    "source": "MedlinePlus (U.S. National Library of Medicine)"
}

# Save to JSON
with open("diabetes_medlineplus.json", "w", encoding="utf-8") as f:
    json.dump(knowledge_base, f, indent=4, ensure_ascii=False)

print("✅ Saved diabetes_medlineplus.json successfully")
print(f"Symptoms: {len(symptoms)}, Risk factors: {len(risk_factors)}, Treatments: {len(treatment_options)}")
```

```python
import json

# Load the disease knowledge base
with open("diabetes_medlineplus.json", "r") as f:
    diabetes_data = json.load(f)

# Convert JSON sections into text chunks for embedding
documents = []
for key, values in diabetes_data.items():
    if isinstance(values, list):
        content = f"{key.upper()}: " + ", ".join(values)
        documents.append(content)
    elif isinstance(values, str):
        documents.append(f"{key.upper()}: {values}")

print(documents[:3])  # Preview chunks
```

# LLM Model Selection (gemma-2b):

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings

# Create embeddings
embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Create and persist the vector store
vectorstore = Chroma.from_texts(documents, embedding_model, collection_name="diabetes_db")
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
from transformers import pipeline

llm = pipeline(
    "text-generation",
    model="google/gemma-2b-it",
    torch_dtype="auto",
    device_map="auto"
)
```

# Prompt Design:

```python
def diagnose(query, prompt_type="zero"):
    # Step 1: Retrieve context
    context_docs = retriever.get_relevant_documents(query)
    context_text = "\n".join([doc.page_content for doc in context_docs])

    # Step 2: Create prompt
    if prompt_type == "zero":
        prompt = f"""
You are a medical diagnostic assistant.
Using the context below, analyze the patient's query and provide:
- Probable diagnosis
- Evidence from context
- Recommended treatment

Context:
{context_text}

User query: {query}
"""
    elif prompt_type == "few":
        prompt = f"""
Example 1:
Symptoms: Thirst, Fatigue
Diagnosis: Type 2 Diabetes
Treatment: Lifestyle modification, Metformin

Example 2:
Symptoms: Chest pain, shortness of breath
Diagnosis: Heart disease
Treatment: Statins, diet control

Now analyze:
Context:
{context_text}

User query: {query}
"""
    elif prompt_type == "meta":
        prompt = f"""
You are a diagnostic reasoning model. Follow these steps:
1. Identify key symptoms.
2. Match them with diseases in context.
3. Justify your choice.
4. Suggest treatment plan.

Context:
{context_text}

User query: {query}
"""
    else:
        raise ValueError("prompt_type must be one of: zero, few, meta")

    # Step 3: Run model
    response = llm(prompt, max_new_tokens=300, temperature=0.4)
    return response[0]['generated_text']
```

```python
query = "The patient has frequent urination, fatigue, and blurred vision."
print(diagnose(query, prompt_type="zero"))
```

```python
print(diagnose(query, prompt_type="few"))
print(diagnose(query, prompt_type="meta"))
```

# Response Generation

```python
synthetic_cases = [
    # Positive (Type 2 Diabetes)
    {"id": 1, "symptoms": "frequent urination, increased thirst, fatigue", "label": "Diabetes"},
    {"id": 2, "symptoms": "blurred vision, slow-healing wounds, weight loss", "label": "Diabetes"},
    {"id": 3, "symptoms": "hunger, tiredness, tingling in hands and feet", "label": "Diabetes"},
    {"id": 4, "symptoms": "dry mouth, frequent infections, unexplained fatigue", "label": "Diabetes"},
    {"id": 5, "symptoms": "numbness in legs, increased appetite, dehydration", "label": "Diabetes"},

    # Negative (Non-diabetic)
    {"id": 6, "symptoms": "runny nose, sore throat, coughing", "label": "Common Cold"},
    {"id": 7, "symptoms": "chest pain, shortness of breath, sweating", "label": "Heart Disease"},
    {"id": 8, "symptoms": "stomach pain, nausea, vomiting", "label": "Gastroenteritis"},
    {"id": 9, "symptoms": "headache, dizziness, light sensitivity", "label": "Migraine"},
    {"id": 10, "symptoms": "joint pain, rash, fever", "label": "Lupus"}
]
```

```python
results = []

for case in synthetic_cases:
    for prompt_type in ["zero", "few", "meta"]:
        response = diagnose(case["symptoms"], prompt_type=prompt_type)
        results.append({
            "id": case["id"],
            "label": case["label"],
            "prompt_type": prompt_type,
            "symptoms": case["symptoms"],
            "response": response
        })
```

```python
import pandas as pd

df_results = pd.DataFrame(results)
df_results.to_csv("diagnostic_responses.csv", index=False)
```

# Evaluate Generated Responses:

```python
from evaluate import load
import nltk
nltk.download('wordnet')
```

```python
perplexity = load("perplexity", module_type="metric")
bleu = load("bleu")
rouge = load("rouge")
meteor = load("meteor")
```

```python
references = [case["label"] for case in synthetic_cases for _ in range(3)]  # repeated for each prompt type
predictions = df_results["response"].tolist()
```

```python
ppl = perplexity.compute(model_id="gpt2", predictions=predictions)
print("Average Perplexity:", ppl["mean_perplexity"])
```

```python
print("BLEU:", bleu.compute(predictions=predictions, references=references))
print("ROUGE:", rouge.compute(predictions=predictions, references=references))
print("METEOR:", meteor.compute(predictions=predictions, references=references))
```

```python
def distinct_n(sentences, n=2):
    ngrams = set()
    total = 0
    for sent in sentences:
        tokens = sent.split()
        total += max(len(tokens) - n + 1, 0)
        ngrams.update(tuple(tokens[i:i+n]) for i in range(len(tokens)-n+1))
    return len(ngrams) / total if total > 0 else 0

print("Distinct-1:", distinct_n(predictions, 1))
print("Distinct-2:", distinct_n(predictions, 2))
```

```python
for prompt_type in ["zero", "few", "meta"]:
    subset = df_results[df_results["prompt_type"] == prompt_type]
    preds = subset["response"].tolist()
    refs = subset["label"].tolist()
    bleu_score = bleu.compute(predictions=preds, references=refs)["bleu"]
    print(f"{prompt_type.upper()} BLEU Score:", bleu_score)
```

- All prompting methods produced correct diagnoses.

- Meta prompting gave the most organized and easy-to-understand reasoning.
